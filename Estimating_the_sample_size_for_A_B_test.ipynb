```python
import numpy as np
import statsmodels.stats.api as sms
from statsmodels.stats.proportion import proportion_effectsize


baseline_conversion = 0.10
relative_lift = 0.03

variant_conversion = baseline_conversion * (1 + relative_lift)

alpha = 0.05
power = 0.80
effect_size = proportion_effectsize(baseline_conversion, variant_conversion)
analysis = sms.NormalIndPower()
sample_size = analysis.solve_power(effect_size=effect_size, alpha=alpha, power=power, ratio=1)
sample_size=sms.NormalIndPower().solve_power(effect_size=effect_size,alpha=alpha,power=power,ratio=1)
print(f"The required number of surveyed users in two groups:{int(np.ceil(sample_size))}")
```

Lowering the significance level to 1% resulted in a larger result as expected. The significance level affects the sample size. Therefore, the lower the probability of a false positive, the larger the sample size should be for the control and experimental groups.

```python
baseline_conversion = 0.10
relative_lift = 0.01
variant_conversion = baseline_conversion * (1 + relative_lift)

alpha = 0.01
power = 0.80

effect_size = proportion_effectsize(baseline_conversion, variant_conversion)

analysis = sms.NormalIndPower()
sample_size = analysis.solve_power(
    effect_size=effect_size, alpha=alpha, power=power, ratio=1)

print(f"Required quantity for ONE group: {int(np.ceil(sample_size))}")
```

Therefore, the sample size depends on such parameters as the significance level, the power of the test and the effect size. The lower the significance level, the probability of obtaining false positive results, the larger the sample size is required. Also, the higher the power of the test and the smaller the effect size, the more data is required to detect significant differences between the control and experimental groups.
