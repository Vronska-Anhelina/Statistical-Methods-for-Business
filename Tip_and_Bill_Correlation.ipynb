```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
sns.scatterplot(data=tips,x="total_bill",y="tip",color="darkorange")
plt.title("Correlation total bills VS tips")
plt.xlabel("Total bill")
plt.ylabel("tips")
plt.show()
```

```python
from scipy.stats import pearsonr
print(f"Test statistic and p-value: {pearsonr(tips["total_bill"], tips["tip"])}")
pearson_corr = tips["total_bill"].corr(tips["tip"])
print(f"Pearson Correlation: {pearson_corr}")
```

```python
from scipy.stats import spearmanr
print(f"Test statistic and p-value: {spearmanr(tips["total_bill"], tips["tip"])}")
spearman_corr = tips["total_bill"].corr(tips["tip"])
print(f"Spearman Correlation: {spearman_corr}")
```

```python
fig,axes=plt.subplots(1, 2, figsize=(12,5))

sns.histplot(tips["total_bill"], ax=axes[0],bins=20,color="olive")
axes[0].set_title("distribution Total Bill")

sns.histplot(tips["tip"],ax=axes[1], bins=20, color="gold")
axes[1].set_title("distribution Tip")
```

```python
print(f"Test statistic and p-value: {spearmanr(tips["total_bill"], tips["size"])}")
spearman_corr_size = tips["total_bill"].corr(tips["size"])
print(f"Spearman Correlation for size: {spearman_corr_size}")
```

```python
print(f"Test statistic and p-value: {pearsonr(tips["total_bill"], tips["size"])}")
pearson_corr_size = tips["total_bill"].corr(tips["size"])
print(f"Pearson Correlation for size: {pearson_corr_size}")
```

Conclusion:
A clear positive trend is visible: the larger the check, the larger the tips. In both cases, the p-value is <0.05, which means that the relationship is statistically significant.
Pearson measures a linear relationship. The r value is usually =0.6, the average correlation.
The Spearman value is ≈0.67, which also indicates an average correlation of melons. Both distributions of the variables are not normal: they are asymmetric with a right tail, similar to an exponential distribution. With a normal distribution, all values ​​are grouped around the mean value, and the deviation decreases. Size is a discrete variable, so it is better to use Spearman.
The p-value in both cases is > 0.05, so there is no reason to reject the null hypothesis, the relationship is not proven or statistically significant. Also, in both cases the correlation is +0.59, which indicates a moderate positive correlation.
