```python
A_sales=[16, 25, 18, 19, 22, 34, 20]
average=sum(A_sales)/len(A_sales)
a_saless=sorted(A_sales)
median=a_saless[len(a_saless)//2]
print(f"Avarage sales for A smartphone:{average}")
print(f"Median for A:{median}")
```

```python
B_sales=[23, 20, 36, 12, 35, 54, 36]
average_sales=sum(B_sales)/len(B_sales)
b_saless=sorted(B_sales)
median_b=b_saless[len(b_saless)//2]
print(f"Avarage sales for B smartphone:{average_sales}")
print(f"Median for B:{median_b}")
```

Comparing both models: Model B sells better, judging by the average sales per day, it sells 8 units more. But the sales of Model A are more stable and predictable. Because in the case of Model A, the mean and median are approximately the same, which means that the sample is symmetric. The sales of this model are the most predictable. What can we say about Model B, where the median is larger than the mean, and therefore the sample contains mostly large numbers (larger than the mean).
