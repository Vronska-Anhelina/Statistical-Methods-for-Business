You manage an online store and are testing three different ad display variants for customers (A, B, C). The goal is to determine which one is most effective by the number of ad clicks (CTR — Click-Through Rate).

Perform an ANOVA test on the three different versions of the advertising campaign. Determine whether there is a significant difference between the campaigns.

**Data Description:**

* `Variant`: Ad version (A, B, or C).
* `Impressions`: Number of ad impressions (how many times the ad variant was shown to users).
* `Clicks`: Number of ad clicks (how many users clicked the ad after seeing it).
* `CTR (Click-Through Rate)`: The ratio of clicks to impressions, expressed as a percentage.


```python
import pandas as pd
data = {
    "Variant": ["A", "A", "A", "B", "B", "B", "C", "C", "C", "A", "B", "C", "A", "B", "C"],
    "Impressions": [100, 150, 120, 90, 130, 140, 110, 100, 130, 160, 180, 120, 140, 110, 100],
    "Clicks": [10, 12, 14, 9, 15, 16, 11, 10, 14, 13, 17, 12, 15, 14, 13]
}
df = pd.DataFrame(data)
df["CTR (%)"] = (df["Clicks"] / df["Impressions"]) * 100
print(df)
```

```python
from scipy.stats import f_oneway
import numpy as np

version_a = df[df["Variant"] == "A"]["CTR (%)"]
version_b = df[df["Variant"] == "B"]["CTR (%)"]
version_c = df[df["Variant"] == "C"]["CTR (%)"]
f_stat,p_value=f_oneway(version_a,version_b,version_c)
print(p_value)
if p_value<0.05:
  print("Різниця є статистично значущою")
else:
  print("Різниця не є статичтисно значущою")
```

There is no significant difference in advertising campaigns. Since, p_value=0.32<0.05, which indicates that the difference between the three advertising campaigns is not statistically significant.

```python
print(f"Середнє значення для А версії:{np.mean(version_a)}")
print(f"Середнє значення для B версії:{np.mean(version_b)}")
print(f"Середнє значення для C версії:{np.mean(version_c)}")
```

Here, the most successful campaign is version B, considering the average value. But as mentioned earlier, this difference is not statistically significant.
