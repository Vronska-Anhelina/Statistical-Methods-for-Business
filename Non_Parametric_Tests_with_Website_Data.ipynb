1. You have data on the time spent on two different product pages. Perform a Mann-Whitney test to determine if there is a difference in the median time spent on the pages.

Initial data:
 ```
 page_a_time = [54.97, 48.62, 56.48, 65.23, 47.66, 47.66, 65.79, 57.67, 45.31, 55.43,
                  45.37, 45.34, 52.42, 30.87, 32.75, 44.38, 39.87, 53.14, 40.92, 35.88]
 page_b_time = [72.59, 52.29, 55.81, 37.9,  48.47, 56.33, 41.19, 59.51, 47.79, 51.5,  
                  47.78, 77.23, 54.84, 42.31, 64.87, 40.35, 57.51, 31.48, 39.06, 57.36]
 ```

````python
from scipy.stats import mannwhitneyu
page_a_time = [54.97, 48.62, 56.48, 65.23, 47.66, 47.66, 65.79, 57.67, 45.31, 55.43,
                 45.37, 45.34, 52.42, 30.87, 32.75, 44.38, 39.87, 53.14, 40.92, 35.88]
page_b_time = [72.59, 52.29, 55.81, 37.9,  48.47, 56.33, 41.19, 59.51, 47.79, 51.5,
                 47.78, 77.23, 54.84, 42.31, 64.87, 40.35, 57.51, 31.48, 39.06, 57.36]
stat,p_value=mannwhitneyu(page_a_time,page_b_time)
print(f"p-value:{p_value}")
if p_value<0.05:
  print("The difference is statistically significant.")
else:
  print("The difference is not statistically significant.")
````

**Conclusions:**
Отже, різниця не є статистично значущою.

2. You have three product pages and you want to see if there is a significant difference in the time users spend on these pages. Run a Kruskal-Wallis test and draw conclusions. On which page did users spend the longest?

Initial data:
 ```
 page_a_seconds = [54.97, 48.24, 59.98, 67.24, 46.91, 46.98, 66.48, 54.35, 42.78, 53.37,
                     43.55, 66.30, 62.28, 42.91, 53.96, 43.61, 59.75, 50.14, 44.56, 48.59]
 page_b_seconds = [59.71, 60.18, 43.29, 58.08, 50.91, 64.05, 54.33, 63.28, 68.80, 50.65,
                     64.79, 72.79, 49.77, 65.23, 48.73, 45.54, 54.47, 57.28, 61.83, 50.38]
 page_c_seconds = [73.41, 44.78, 67.57, 62.83, 55.82, 80.32, 43.27, 71.05, 65.68, 43.36,
                     75.98, 46.57, 69.03, 55.65, 85.73, 58.39, 69.28, 56.21, 60.14, 53.19]
 ```

````python
from scipy.stats import kruskal


page_a_seconds = [54.97, 48.24, 59.98, 67.24, 46.91, 46.98, 66.48, 54.35, 42.78, 53.37,
                    43.55, 66.30, 62.28, 42.91, 53.96, 43.61, 59.75, 50.14, 44.56, 48.59]
page_b_seconds = [59.71, 60.18, 43.29, 58.08, 50.91, 64.05, 54.33, 63.28, 68.80, 50.65,
                    64.79, 72.79, 49.77, 65.23, 48.73, 45.54, 54.47, 57.28, 61.83, 50.38]
page_c_seconds = [73.41, 44.78, 67.57, 62.83, 55.82, 80.32, 43.27, 71.05, 65.68, 43.36,
                    75.98, 46.57, 69.03, 55.65, 85.73, 58.39, 69.28, 56.21, 60.14, 53.19]

stat,p_value=kruskal(page_a_seconds,page_b_seconds,page_c_seconds)
print(f"P-value:{p_value}")
if p_value<0.05:
  print("The difference is statistically significant.")
else:
  print("The difference is not statistically significant.")
````

**Conclusions:** In conclusion, the difference between the three pages is statistically significant.

3. You want to compare the effectiveness of two advertising campaigns for the same group of users. Perform a Wilcoxon test and determine if there is a significant difference in conversion before and after changing the advertising strategy.

Initial data:
 ```
 conversion_before = [2.75, 2.49, 2.97, 3.65, 2.41, 2.41, 3.10, 2.89, 2.15, 2.68,
                        2.18, 3.63, 3.39, 2.15, 2.69, 2.16, 3.02, 2.53, 2.26, 2.47]
 conversion_after = [3.13, 3.07, 3.55, 2.92, 3.28, 3.65, 2.69, 2.85, 3.44, 3.16,
                       3.62, 3.97, 2.45, 3.53, 2.65, 2.81, 2.90, 3.19, 3.08, 2.69]
 ```

````python
from scipy.stats import wilcoxon
conversion_before = [2.75, 2.49, 2.97, 3.65, 2.41, 2.41, 3.10, 2.89, 2.15, 2.68,
                       2.18, 3.63, 3.39, 2.15, 2.69, 2.16, 3.02, 2.53, 2.26, 2.47]
conversion_after = [3.13, 3.07, 3.55, 2.92, 3.28, 3.65, 2.69, 2.85, 3.44, 3.16,
                      3.62, 3.97, 2.45, 3.53, 2.65, 2.81, 2.90, 3.19, 3.08, 2.69]

stat,p_value=wilcoxon(conversion_before,conversion_after)
print(f"P-value:{p_value}")
if p_value<0.05:
  print("The difference is statistically significant.")
else:
  print("The difference is not statistically significant.")
````

**Conclusions:** So, the difference between the previous interface and the new one is statistically significant. So, the new version is better.
