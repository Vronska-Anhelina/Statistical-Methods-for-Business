```python
import matplotlib.pyplot as plt
conversion_rate_A = [5.2, 4.8, 6.1, 5.0, 5.3, 4.9, 5.1, 5.4, 4.7, 5.0]
conversion_rate_B = [6.3, 6.7, 6.9, 6.2, 7.0, 6.5, 6.8, 6.6, 6.1, 6.4]

fig,ax=plt.subplots(1,2,figsize=(10,4))
ax[0].hist(conversion_rate_A,bins=5,color="lightblue",alpha=0.7,edgecolor="black")
ax[0].set_title("Distibution for conversion A")
ax[1].hist(conversion_rate_B,bins=5,color="slateblue",alpha=0.7,edgecolor="black")
ax[1].set_title("Distibution for conversion B")
```

One of the distributions is non-normal, so I continue using the Mann-Whitney test.


```python
from scipy.stats import mannwhitneyu
u_stat,p_value=mannwhitneyu(conversion_rate_A,conversion_rate_B)
print(f"U-stats:{u_stat}, p-value={p_value}")
```

In connection with the obtained results, p-value=0.0002<0.005 and this means that the conversions are statistically different. And the result is not random, and therefore, this means that the new loyalty program is effective.

```python
satisfaction_A = [3, 4, 5, 3, 4, 3, 4, 5, 3, 4]
satisfaction_B = [4, 5, 5, 4, 5, 5, 4, 5, 5, 4]
```

We are now working with ordinal data (customer satisfaction levels are determined from 1 to 5), so we will also use the Mann-Whitney test.

```python
u_stat,p_value=mannwhitneyu(satisfaction_A,satisfaction_B)
print(f"U-stats:{u_stat}, p-value={p_value}")
```

In the obtained results, p_value=0.02 which is greater than 0.005, and therefore, the level of customer satisfaction depending on the loyalty program is not statistically significant and does not differ statistically. Both in customers who use the old loyalty program and in customers who use the new one.
