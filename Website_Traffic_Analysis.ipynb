```python
import numpy as np


visitors=[150,200, 180, 220,	190,	250,	170, 1600, 210,	230]
range=max(visitors)-min(visitors)
print(f"Range of visitors:{range}")
```

```python
visitor_sort=sorted(visitors)
visit_left=visitor_sort[: len(visitor_sort)//2]
visit_right=visitor_sort[len(visitor_sort)//2:]
q1=np.median(visit_left)
q2=np.median(visitor_sort)
q3=np.median(visit_right)

print(f"Q1:{q1}")
print(f"Median(Q2):{q2}")
print(f"Q3:{q3}")
```

```python
IQR=q3-q1
print(IQR)
```

```python
std_dev = np.std(visitor_sort)
print(f"standard deviation:{std_dev}")
```

Using the calculations above, we can say that the data is quite variable. This is indicated by the large value of the range. It also indicates instability in site visits during a given period. The larger the range indicator, the greater the instability that can be seen in visits. The interquartile range is 50, which indicates a narrow range. Comparing it with the range of the entire data set and the interquartile range, we can say that there are anomalous values ​​in the set (in our case 1600), which affect the calculation of the range (the range is sensitive to outliers, unlike the interquartile range). The typical value of the number of visitors is the median of this sample. That is, 205 visitors. The standard deviation shows that most of the values ​​​​variate by 420 units from the mean value. This indicates extreme values ​​​​(outliers) in the set and the sensitivity of the standard deviation to such values.
