```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data={"sales":[5, 15, 7, 9, 2, 10, 11, 25, 7, 8, 2, 3, 10, 10, 8]}
sales_data=pd.DataFrame(data)
def bootstrap(sales_data,n_iterations):
  sample_means=[]
  for i in range(n_iterations):
    sample=sales_data.sample(frac=1,replace=True)
    sample_means.append(sample['sales'].mean())

  return sample_means
bootstrap_mean= bootstrap(sales_data,1000)
plt.hist(bootstrap_mean,bins=5,color="red")
plt.title("Bootstrap Distribution of Sample Means")
plt.xlabel("Median")
plt.ylabel("Frequency")
plt.show()
```

The most likely median value is 7-8. The distribution is normal, centered around the mean.
